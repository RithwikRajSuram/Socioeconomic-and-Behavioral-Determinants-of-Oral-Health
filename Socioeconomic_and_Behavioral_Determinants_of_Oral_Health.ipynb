```python
import matplotlib.pyplot as plt
import numpy as np

# SLU colors
slu_dark_blue = '#003DA5'
slu_mid_blue  = '#4F81BD'
slu_light_blue = '#A6C8FF'
slu_gray = '#D9D9D9'

# -------------------------------
# Figure 1 – Bar Chart
# -------------------------------
income_levels = ['Low Income', 'Middle Income', 'High Income']
prevalence = [34, 24, 15]

plt.figure(figsize=(7,5))
bars = plt.bar(income_levels, prevalence,
               color=[slu_dark_blue, slu_mid_blue, slu_light_blue])
plt.title('Figure 1. Prevalence of Untreated Dental Caries by Income Level (NHANES 2013–2018)',
          fontsize=12, pad=15, color=slu_dark_blue)
plt.ylabel('Prevalence (%)')
plt.ylim(0,40)
plt.grid(axis='y', linestyle='--', color=slu_gray, alpha=0.7)
for bar,val in zip(bars, prevalence):
    plt.text(bar.get_x()+bar.get_width()/2, val+1, f'{val}%',
             ha='center', fontsize=10)
plt.tight_layout()
plt.show()

# -------------------------------
# Figure 2 – Forest Plot
# -------------------------------
predictors = ['Income-to-Poverty Ratio', 'Education (College vs HS)', 'Current Smoker']
odds_ratios = [0.74, 0.69, 1.95]
lower_ci = [0.61, 0.55, 1.60]
upper_ci = [0.89, 0.86, 2.38]
y_pos = np.arange(len(predictors))

plt.figure(figsize=(8,5))
plt.errorbar(odds_ratios, y_pos,
             xerr=[np.subtract(odds_ratios, lower_ci),
                   np.subtract(upper_ci, odds_ratios)],
             fmt='o', color=slu_dark_blue, ecolor='black',
             elinewidth=1.5, capsize=5)
plt.axvline(x=1, color=slu_gray, linestyle='--')
plt.yticks(y_pos, predictors)
plt.xlabel('Odds Ratio (95% CI)')
plt.title('Figure 2. Adjusted Odds Ratios for Socioeconomic and Behavioral Predictors of Poor Oral Health',
          fontsize=11, pad=15, color=slu_dark_blue)
plt.xlim(0.4,2.5)
plt.grid(axis='x', linestyle='--', color=slu_gray, alpha=0.7)
plt.tight_layout()
plt.show()
```
